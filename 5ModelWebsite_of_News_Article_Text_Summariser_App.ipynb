```python
from flask import Flask,render_template, request
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)
 
class summary_generation:
  def __init__(self,body,modelname,result):
    self.body=body
    self.modelname=modelname
    self.result=result
  
  def bertSum(self):
    from summarizer import Summarizer
    bert_model = Summarizer()
    return ''.join(bert_model(self.body, min_length=60))
 
  def gptSum(self):
    !pip install transformers==2.2.0
    from summarizer import TransformerSummarizer
    GPT2_model = TransformerSummarizer(transformer_type="GPT2",transformer_model_key="gpt2-medium")
    return ''.join(GPT2_model(self.body, min_length=60))
 
  def xlnetSum(self):
    !pip install transformers==2.2.0
    from summarizer import TransformerSummarizer
    xlnet_model = TransformerSummarizer(transformer_type="XLNet",transformer_model_key="xlnet-base-cased")
    return ''.join(xlnet_model(self.body, min_length=60))

  def bartSum(self):
    !pip install transformers==4.5.0
    from transformers import pipeline
    summarizer = pipeline('summarization', model='facebook/bart-large-cnn', tokenizer='facebook/bart-large-cnn')
    summary=summarizer(self.body, min_length = round(0.1 * len(self.body.split(' '))), max_length = round(0.2 * len(self.body.split(' '))), do_sample=False)
    return summary[0]['summary_text']
  
  def t5Sum(self):
    !pip install transformers==4.5.0
    !pip install sentencepiece
    from transformers import T5Tokenizer, T5ForConditionalGeneration
    model = T5ForConditionalGeneration.from_pretrained('t5-base')
    tokenizer = T5Tokenizer.from_pretrained('t5-base')
    tokens_input = tokenizer.encode(self.body, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(tokens_input, min_length=60, max_length=180, length_penalty=4.0)
    summary = tokenizer.decode(summary_ids[0])
    summary = re.sub(r'<pad>','',summary)
    summary = re.sub(r'<extra_id_0>','',summary)
    summary = re.sub(r'</s>','',summary)
    return summary
  
  def set_model(self,mname):
    self.modelname=mname
 
  def summarise(self):
    if self.modelname == 'GPT-2':
      self.result = self.gptSum()
    elif self.modelname == 'BERT':
      self.result = self.bertSum()
    elif self.modelname == 'XLNet':
     self.result = self.xlnetSum()
    elif self.modelname == 'BART':
     self.result = self.bartSum()
    elif self.modelname == 'T5':
     self.result = self.t5Sum()
 
SG=summary_generation('','GPT-2','');
finalsummary=''
mname=''
 
@app.route("/",methods=['GET','POST'])
def home():
  if request.method== 'POST':
    mname=request.form.get('mname')
    SG.set_model(mname)
  return render_template('home.html')
 
@app.route("/enter-news-article",methods=['GET','POST'])
def input():
  if request.method== 'POST':
    SG.body=request.form['newsbody']
    SG.summarise()
  return render_template('newsarticle.html',mname=SG.modelname)
 
@app.route("/summary",methods=['GET','POST'])
def summary():
  return render_template('summary.html',finalsummary=SG.result)
 
app.run()
```
